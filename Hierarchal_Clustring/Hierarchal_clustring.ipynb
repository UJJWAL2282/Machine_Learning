```python
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
#Importing  iris the dataset
iris=datasets.load_iris()
iris
```

```python
iris_data=pd.DataFrame(iris.data)
iris_data.head()
```

```python
iris_data.columns=iris.feature_names
iris_data.head()
```

```python
# Standardizing the data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaler=scaler.fit_transform(iris_data)
X_scaler
```

```python
#Apply PCA To reduce the dimensions

from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca_scaled=pca.fit_transform(X_scaler)
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=iris.target)
```

```python
#Agglomerative Clustring
import scipy.cluster.hierarchy as sch
##plot the dendrogram
dendrogram=sch.dendrogram(sch.linkage(pca_scaled,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Eculidian Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
cluster.fit(pca_scaled)
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cluster.labels_)
```
